```python
def metrica(cadena):
    vocales = ['a', 'e', 'i', 'o', 'u']
    vocal_completa = ['a', 'e', 'i', 'o', 'u', 'á', 'é', 'í', 'ó', 'ú']
    vocal_acento = ['á', 'é', 'í', 'ó', 'ú']
    consonantes = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'ñ', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'y', 'z']
    silabas = []
    monosilabas = []
    agudas_lista = []
    graves = []
    esdrujulas_lista = []
    clasificador = []
    letra_anterior = ''


    cadena_separada = cadena.split()
    print(cadena_separada)

    # para las monosilabas
    # tienen 1, 2 o 3 letras
    for palabra in cadena_separada[:]:
        if len(palabra) <= 3:
            clasificador.append(palabra)
            monosilabas.append(palabra)
            cadena_separada.remove(palabra)

    print("Monosilabas =", monosilabas)
    print(cadena_separada)

   

    # para las vocal_completa (tienen o no tilde)
    # tienen tilde en la ultima vocal terminan en vocal, n o s
    # en toda la palabra no hay tilde y las ultimas dos sean vocal + consonante
    for palabra in cadena_separada[:]:
        if (palabra[-1] in ['n','s'] and palabra[-2] in vocal_acento) \
        or palabra[-1] in vocal_acento \
        or (palabra[-1] in consonantes and palabra[-2] in vocales and any(letra not in vocal_acento for letra in palabra)):
            clasificador.append(palabra)
            agudas_lista.append(palabra)
            cadena_separada.remove(palabra)

    print("aguda =", agudas_lista)
    print(cadena_separada)

    # para las llanas o graves
    # NO terminan en n, s o vocal
    for palabra in cadena_separada[:]:
        if (palabra[-1] in ['n','s'] and palabra[-2] in vocales):
            clasificador.append(palabra)
            graves.append(palabra)
            cadena_separada.remove(palabra)

    print("Graves =", graves)
    print(cadena_separada)

   # para las esdrújulas
    # siempre llevan tilde
    for palabra in cadena_separada[:]:
        if any(letra in vocal_acento for letra in palabra):
            clasificador.append(palabra)
            esdrujulas_lista.append(palabra)
            cadena_separada.remove(palabra)

    print("vocal_acento =", esdrujulas_lista)
    print(cadena_separada)

    print("clasificador =", clasificador)



    for letra in cadena:
        print("iteracion")
        indice_actual = cadena.find(letra)
        print(letra, indice_actual)
        if letra in vocal_completa:
            if indice_actual != 0:
                letra_anterior = cadena[indice_actual - 1]

            if letra_anterior in vocal_completa and letra in vocal_completa:
                if indice_actual + 1 < len(cadena):
                    silabas[-1] += letra

            elif letra in vocal_completa:
                silaba = cadena[:indice_actual + 1]
                silabas.append(silaba)
                cadena = cadena.split(silaba)
                cadena = cadena[1]

    return silabas


metrica('patatas')

```

```python
class char():
    def __init__(self):
        pass
    
class char_line():
    def __init__(self, word):
        self.word = word
        self.char_line = [(char, self.char_type(char)) for char in word]
        self.type_line = ''.join(chartype for char, chartype in self.char_line)
        
    def char_type(self, char):
        if char in set(['a', 'á', 'e', 'é','o', 'ó', 'í', 'ú']):
            return 'V' #strong vowel
        if char in set(['i', 'u', 'ü']):
            return 'v' #week vowel
        if char=='x':
            return 'x'
        if char=='s':
            return 's'
        else:
            return 'c'
            
    def find(self, finder):
        return self.type_line.find(finder)
        
    def split(self, pos, where):
        return char_line(self.word[0:pos+where]), char_line(self.word[pos+where:])
    
    def split_by(self, finder, where):
        split_point = self.find(finder)
        if split_point!=-1:
            chl1, chl2 = self.split(split_point, where)
            return chl1, chl2
        return self, False
     
    def __str__(self):
        return self.word
    
    def __repr__(self):
        return repr(self.word)

class silabizer():
    def __init__(self):
        self.grammar = []
        
    def split(self, chars):
        rules  = [('VV',1), ('cccc',2), ('xcc',1), ('ccx',2), ('csc',2), ('xc',1), ('cc',1), ('vcc',2), ('Vcc',2), ('sc',1), ('cs',1),('Vc',1), ('vc',1), ('Vs',1), ('vs',1)]
        for split_rule, where in rules:
            first, second = chars.split_by(split_rule,where)
            if second:
                if first.type_line in set(['c','s','x','cs']) or second.type_line in set(['c','s','x','cs']):
                    #print 'skip1', first.word, second.word, split_rule, chars.type_line
                    continue
                if first.type_line[-1]=='c' and second.word[0] in set(['l','r']):
                    continue
                if first.word[-1]=='l' and second.word[-1]=='l':
                    continue
                if first.word[-1]=='r' and second.word[-1]=='r':
                    continue
                if first.word[-1]=='c' and second.word[-1]=='h':
                    continue
                return self.split(first)+self.split(second)
        return [chars]
        
    def __call__(self, word):
        return self.split(char_line(word))


var = char_line('en todo el mar conocido ratón jugar rubí')
var.split()
```
