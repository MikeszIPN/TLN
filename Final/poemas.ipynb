```python
def metrica(cadena):
    vocales = ['a', 'e', 'i', 'o', 'u']
    agudas = ['a', 'e', 'i', 'o', 'u', 'n', 's']
    esdrujulas = ['á','é','í','ó','ú']
    silabas = []
    monosilabas = []
    agudas = []
    graves = []
    esdrujulas = []
    letra_anterior = ''

    cadena_separada = cadena.split()
    print(cadena_separada)

    # para las monosilabas
    # tienen 1, 2 o 3 letras
    for palabra in cadena_separada[:]:
        if len(palabra) <= 3:
            silabas.append(palabra)
            monosilabas.append(palabra)
            cadena_separada.remove(palabra)

    print("Monosilabas =", monosilabas)
    print(cadena_separada)

    # para las agudas
    # terminan en vocal, n o s
    for palabra in cadena_separada[:]:
        if palabra[-1] in agudas:
            silabas.append(palabra)
            agudas.append(palabra)
            cadena_separada.remove(palabra)

    print("Agudas =", agudas)
    print(cadena_separada)

    # para las llanas o graves
    # NO terminan en n, s o vocal
    for palabra in cadena_separada[:]:
        if palabra[-1] not in agudas:
            silabas.append(palabra)
            graves.append(palabra)
            cadena_separada.remove(palabra)

    print("Graves =", graves)
    print(cadena_separada)

    # para las esdrújulas
    # al menos una letra lleva tilde
    for palabra in cadena_separada[:]:
        if any(letra in esdrujulas for letra in palabra):
            silabas.append(palabra)
            esdrujulas.append(palabra)
            cadena_separada.remove(palabra)

    print("Esdrujulas =", esdrujulas)
    print(cadena_separada)

    print("Silabas =", silabas)

    for letra in cadena:
        indice_actual = cadena.index(letra)

        if letra in vocales:
            if indice_actual != 0:
                letra_anterior = cadena[indice_actual - 1]

            if letra_anterior in vocales and letra in vocales:
                if indice_actual + 1 < len(cadena):
                    silabas[-1] += letra

            elif letra in vocales:
                silaba = cadena[:indice_actual + 1]
                silabas.append(silaba)
                cadena = cadena.split(silaba, 1)
                # print(cadena, silaba)  
                cadena = cadena[1]

    return silabas


metrica('en todo el mar conocido música')
```
