Lee el archivo en txt y lo separa en lineas

```python
file_name = "prueba.txt" 
file_name_mod = file_name.replace(".txt", "") # Elim
file_name_mod = file_name_mod + "m.txt" # Nombre del

mod_file = open("../Final/" + file_name_mod, 'w', encoding='utf-8') # Creo el acrchi
raw_file = open("../Final/" + file_name, 'r', encoding='utf-8') # Abro el archivo pa

for x in raw_file: 
    xf = x.replace(".", "")
    xr = xf.replace(",", "")
    xs = xr.replace("(", "")
    xt = xs.replace(")", "")
    xu = xt.replace(":", "")
    xv = xu.replace(";", "")
    xw = xv.replace("¿", "")
    xx = xw.replace("?", "")
    xy = xx.replace("-", "")
    xz = xy.replace("!", "")
    xa = xz.replace("¡", "")
    xb = xa.replace("\"", "")
    xc = xb.replace("—", "")

    mod_file.write(xc)

mod_file.close()
raw_file.close()

# Contar lineas sin espacios y guardarlas--------
archivo = open("../Final/pruebam.txt", encoding='utf-8')

Content = archivo.read()
CoList = Content.split("\n")

for i in range (CoList.count('')):
    CoList.remove('')

print(CoList)
```

obtiene las silabas de las palabras del texto

```python
from pylabeador import syllabify

resultado = []
palabras = []
finales = []
suma = 0
for line in CoList:
    text = line.split()
   
    #print(text)
    finales.append(text[-1])    # guarda las ultimas palabras_finales de cada linea

    for word in text:   # separa las palabras_finales de cada oracion
        palabras.append(word)
        
    
for palabra in palabras:    # obtiene las silabas de las palabras de las oraciones
    resultado.append(syllabify(palabra))


#print(resultado)
print(finales)

# cuenta las silabas del poema 
for i in resultado:
    suma += len(i)
    #print(i)

print(suma)
```

```python
def tipo_acentuacion(palabras_finales, silabas):
     vocales = ['a', 'e', 'i', 'o', 'u']
     vocal_acento = ['á', 'é', 'í', 'ó', 'ú']
     consonantes = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'ñ', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'y', 'z']
     monosilabas = []
     agudas = []
     graves = []
     esdrujulas = []
     clasificador = []

     print(palabras_finales)
     print(silabas)

    
     for palabra in palabras_finales[:]:
          # para las monosilabas
          # tienen 1, 2 o 3 letras
          if len(palabra) <= 3:
               clasificador.append("monosilaba")
               monosilabas.append(palabra)
               palabras_finales.remove(palabra)
               continue
               # return 1
          
          # para las agudas (tienen o no tilde)
          # tienen tilde en la ultima vocal terminan en vocal, n o s
          # en toda la palabra no hay tilde y las ultimas dos sean vocal + consonante
          # for palabra in palabras_finales[:]:
          # las agudas tienen acento en la ultima silaba
          if silabas[-1] in vocal_acento and (palabra[-1] in ['n','s'] or palabra[-1] in vocales):
               clasificador.append("aguda")
               agudas.append(palabra)
               palabras_finales.remove(palabra)
               continue

          if (palabra[-1] not in ['n','s'] ) \
               or palabra[-1] == 'y' \
               or palabra[-1] in vocales \
               or (palabra[-1] in vocal_acento):

               clasificador.append("aguda")
               agudas.append(palabra)
               palabras_finales.remove(palabra)
               continue
               # return 1
          
          # para las llanas o graves
          # NO terminan en n, s o vocal
          #for palabra in silabas[:]:
          # llevan tilde en la penultima silaba
          if silabas[-2] in vocal_acento and palabra[-1] not in ['n','s']:
               clasificador.append("grave")
               graves.append(palabra)
               palabras_finales.remove(palabra)
               continue

          if palabra[-1] in vocales or palabra[-1] in ['n','s'] or palabra[-1] in vocales:
               clasificador.append("grave")
               graves.append(palabra)
               palabras_finales.remove(palabra)
               continue
               # return 0
          
          # para las esdrújulas
          # siempre llevan tilde
          #for palabra in silabas[:]:
          if any(palabra in vocal_acento for palabra in palabras):
               clasificador.append("esdrujula")
               esdrujulas.append(palabra)
               palabras_finales.remove(palabra)
               continue
               # return -1
          
     print("Monosilabas =", monosilabas)
     print("Agudas =", agudas)
     print("Graves =", graves)
     print("Esdrujulas =", esdrujulas)
     print("palabras_finales = ", palabras_finales)

     print("clasificador =", clasificador)    
```
