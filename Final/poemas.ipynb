```python
def signos_puntuacion(texto):
    puntuacion = ".,;:-¿?¡!()"

    for char in puntuacion:
        texto = texto.replace(char, " ")

    return texto

def normalizacion(file_name):
    file_name_mod = file_name.replace(".txt", "") # Elim
    file_name_mod = file_name_mod + "m.txt" # Nombre del

    mod_file = open("" + file_name_mod, 'w', encoding='utf-8') # creo el archi
    raw_file = open("" + file_name, 'r', encoding='utf-8')  # abro el archivo pa

    for line in raw_file:
        # elimina varios signos de puntuación
        line = signos_puntuacion(line)

        # para eliminar varios signos seguidos
        words = line.split()
        cleaned_line = ' '.join(words)

        mod_file.write(cleaned_line + "\n")

    mod_file.close()
    raw_file.close()

    archivo = open(file_name_mod, encoding='utf-8')

    # Contar líneas sin espacios y guardarlas
    Content = archivo.read()
    CoList = Content.split("\n")

    for i in range(CoList.count('')):
        CoList.remove('')

    archivo.close()

    return CoList

texto = normalizacion('prueba.txt')
print(texto[0])
```

```python
def tipo_acentuacion(cadena):
     vocales = ['a', 'e', 'i', 'o', 'u']
     vocal_acento = ['á', 'é', 'í', 'ó', 'ú']
     consonantes = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'ñ', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'y', 'z']
     monosilabas = []
     agudas_lista = []
     graves = []
     esdrujulas_lista = []
     clasificador = []

     cadena_separada = cadena.split()
     print(cadena_separada)

     # para las monosilabas
     # tienen 1, 2 o 3 letras
     for palabra in cadena_separada[:]:
          if len(palabra) <= 3:
               clasificador.append(palabra)
               monosilabas.append(palabra)
               cadena_separada.remove(palabra)

     print("Monosilabas =", monosilabas)
     print(cadena_separada)

     
     # para las vocal_completa (tienen o no tilde)
     # tienen tilde en la ultima vocal terminan en vocal, n o s
     # en toda la palabra no hay tilde y las ultimas dos sean vocal + consonante
     for palabra in cadena_separada[:]:
          if (palabra[-1] in ['n','s'] and palabra[-2] in vocal_acento) \
          or palabra[-1] in vocal_acento \
          or (palabra[-1] in consonantes and palabra[-2] in vocales and any(palabra not in vocal_acento for palabra in palabra)):
               clasificador.append(palabra)
               agudas_lista.append(palabra)
               cadena_separada.remove(palabra)

     print("aguda =", agudas_lista)
     print(cadena_separada)

     # para las llanas o graves
     # NO terminan en n, s o vocal
     for palabra in cadena_separada[:]:
          if (palabra[-1] in ['n','s'] and palabra[-2] in vocales):
               clasificador.append(palabra)
               graves.append(palabra)
               cadena_separada.remove(palabra)

     print("Graves =", graves)
     print(cadena_separada)

     # para las esdrújulas
     # siempre llevan tilde
     for palabra in cadena_separada[:]:
          if any(palabra in vocal_acento for palabra in palabra):
               clasificador.append(palabra)
               esdrujulas_lista.append(palabra)
               cadena_separada.remove(palabra)

     print("vocal_acento =", esdrujulas_lista)
     print(cadena_separada)

     print("clasificador =", clasificador)    
```

```python
def metrica(cadena):
    cadena = cadena + ' '
    vocal = ['a', 'e', 'i', 'o', 'u']
    vocal_acento = ['á', 'é', 'í', 'ó', 'ú']
    vocal_completa = ['a', 'e', 'i', 'o', 'u', 'á', 'é', 'í', 'ó', 'ú']
    consonantes = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'ñ', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'y', 'z']
    consonantes_seguidas = ['pr', 'dr', 'br', 'cr', 'fr', 'gr', 'kr', 'tr', 'fl', 'kl', 'cl', 'bl', 'pl', 'gl', 'ch', 'll']

    silabas = []
    silaba = ''
 
    palabras = cadena.split()

    for palabra in palabras:
        print(palabra)

        # filtra si es monosilaba
        if len(palabra) <= 3:
            silabas.append(palabra)
        else:
            composicion = []

            # determina que tipo de letra tiene
            for i in range(len(palabra)):
                if palabra[i] in vocal:
                   composicion.append("v") 
                if palabra[i] in vocal_acento:
                    composicion.append("va")
                if palabra[i] in consonantes:
                    composicion.append("c")
            print(composicion) 
            # 3) Cuando existan dos consonantes seguidas de la siguiente lista:  
            # pr, dr, br, cr, fr, gr, kr, tr, fl, kl, cl, bl, pl, gl, ch y ll
            # permanecerán juntas en la misma sílaba junto a la vocal que les sigue 
            i = 0
            while i < len(composicion):
                if palabra[i:i+2].lower() in consonantes_seguidas:
                    if i + 3 < len(composicion) and composicion[i+3] == 'c' and palabra[i+3] != 'x':
                        # Las dos consonantes seguidas se mantienen en la misma sílaba con la vocal que les sigue
                        silaba = palabra[i:i+4]
                        silabas.append(silaba)
                        i += 4  # Avanzamos 4 posiciones (2 consonantes + 2 vocales)
                    else:
                        # Las dos consonantes seguidas se mantienen en la misma sílaba con la vocal que les sigue
                        silaba = palabra[i:i+3]
                        silabas.append(silaba)
                        i += 3  # Avanzamos 3 posiciones (2 consonantes + 1 vocal)
                # 2) Cuando hay una consonante entre dos vocales, la consonante forma parte de la sílaba de la vocal siguiente
                elif composicion[i:i+3] == ['c', 'v', 'c']:
                    silaba = palabra[i:i+2]
                    silabas.append(silaba)
                    i += 3  # Avanzamos 3 posiciones (2 consonantes + 1 vocal)
                # 1) Toda sílaba lleva necesariamente una vocal pero puede o no tener consonantes (creo que aqui van los diptongos)
                elif composicion[i:i+2] == ['c','v']:
                    silaba = palabra[i:i+1]
                    silabas.append(silaba)
                    i += 2  # Avanzamos 3 posiciones (2 consonantes + 1 vocal)
                else:
                    # Tratamos una sola letra (vocal o consonante) como una sílaba
                    silabas.append(palabra[i])
                    i += 1
                
            # Manejar la última sílaba
            if composicion[-3:] == ['v', 'c', 'v']:
                ultima_silaba = palabra[-2:]
                silabas.append(ultima_silaba)
   

            print(silabas)
    return silabas


resultado = metrica("Triste graNde cRisto aorta águla")
print(resultado)
```
